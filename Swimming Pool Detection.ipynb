# Swimming Pool Detection Using Python & ArGIS API

In this project, I will prepare our sample data by splitting it into training and validation sets, applying data augmentation techniques, and organizing it into batches for efficient memory management. I'll utilize the ArcGIS API's **prepare_data** method to automate this process, which includes default transformations for data augmentation.

#### **Importing the required packages**

```python
from arcgis.learn import SingleShotDetector, prepare_data
```

Loading the image chips into memory and running the **prepare_data** method

```python
data_path = r'C:\Users\Moses Ogbonna\ObjectDetection\ImageChips' 

data = prepare_data(data_path, {0:'Pool'})
data.classes
```

### Visualizing the augmented training sample data

```python
data.show_batch()
```

### Defining the deep learning model architecture

```python
ssd = SingleShotDetector(data)
```

### Finding an optimal learning rate

```python
ssd.lr_find()
```

Based on the learning rate plot above, the loss decreases dramatically at 1e-3 (0.001), so I will use this rate for training.

#### **Training the SSD model for 25 epochs, using a learning rate of 0.01**

```python
ssd.fit(25, lr=1e-3)
```

### Reviewing the model

```python
ssd.show_results()
```

#### Plotting the training and validation losses

```python
ssd.plot_losses()
```

#### Printing the average precision score of the model

```python
ssd.average_precision_score()
```

    This model provides a good baseline, predicting most of the pools identified in the ground truth. We could modify the parameters to improve the results of the model. 

### Saving the model

```python
ssd.save(r'C:\Users\Mosses Ogbonna\Documents\Pools_Model',save_inference_file=True)
```
